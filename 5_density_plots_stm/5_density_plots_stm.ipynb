There is a file in the on-line resources called stm. txt, which contains a grid of values from scanning tunneling microscope measurements of the (111} surface
of silicon. A scanning tunneling microscope (STM) is a device that measures the shape of a surface at the atomic level by tracking a sharp tip over the surface and measuring
quantum tunneling current as a function of position. The end result is a grid of values that represent the height of the surface and the file stm. txt contains just such a grid of
values. Write a program that reads the data contained in the file and makes a density plot of the values. Use the various options and variants you have learned about to make
a picture that shows the structure of the silicon surface clearly.

**Answer:**

```python
# Need to import loadtxt function from numpy module

data = loadtxt("stm.txt", float)

print("\nIllustration of structure of the silicon surface using density plots from STM measurements.\n")
# Using default colorscheme
print("\nDefault colorscheme: 'jet'\n")
shw = plt.imshow(data, origin="lower", extent=[0,600,0,600])
bar = plt.colorbar(shw)
bar.set_label('Colorbar')
plt.title('\nSTM Measurements of (111) Silicon Surface\n')
show()





# Using 'hot' colorscheme
print("\nColorscheme: 'hot'\n")
shw1 = plt.imshow(data,cmap='hot', origin="lower", extent=[0,600,0,600])
bar1 = plt.colorbar(shw1)
bar1.set_label('Colorbar')
plt.title('\nSTM Measurements of (111) Silicon Surface\n')
show()

# Using 'hsv' corlorscheme
print("\nColorscheme: 'hsv'\n")
shw2 = plt.imshow(data,cmap='hsv', origin="lower", extent=[0,900,0,600], aspect=1.2)
bar2 = plt.colorbar(shw2)
bar2.set_label('Colorbar')
plt.title('\nSTM Measurements of (111) Silicon Surface\n')
show()


# Using 'gray' corlorscheme
print("\nColorscheme: 'gray'\n")
shw3 = plt.imshow(data,cmap='gray', origin="lower", extent=[0,800,0,800])
bar3 = plt.colorbar(shw3)
bar3.set_label('Colorbar')
plt.title('\nSTM Measurements of (111) Silicon Surface\n')
show()
```
